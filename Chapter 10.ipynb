# 資料分組/樞紐分析表
## 資料分組
### 分組鍵是欄名

```python
import pandas as pd
df = pd.read_excel("C:/ACD019600/data/test10.xlsx")
df
```

```python
df.groupby("客戶分類")
```

```python
df.groupby("客戶分類").count()
```

```python
df.groupby("客戶分類").sum()
```

```python
#對分組後的資料進行計數運算
df.groupby(["客戶分類","區域"]).count()
```

```python
#對分組後的資料進行求和運算
df.groupby(["客戶分類","區域"]).sum()
```

```python
df.groupby("客戶分類")["使用者ID"].count()
```

### 分組鍵是Series

```python
df["客戶分類"]
```

```python
#對分組以後的資料進行計數運算
df.groupby(df["客戶分類"]).count()
```

```python
#對分組以後的資料進行求和運算
df.groupby([df["客戶分類"],df["區域"]]).sum()
```

### 神奇的aggregate方法

```python
df = pd.read_excel("C:/ACD019600/data/test10.xlsx", sheet_name="工作表2")
df
```

```python
df.groupby("客戶分類").aggregate(["count","sum"])
```

```python
df.groupby("客戶分類").aggregate({"使用者ID":"count","7月銷量":"sum","8月銷量":"sum"})
```

### 對分組後的結果重置索引

```python
df.groupby("客戶分類").sum()
```

```python
df.groupby("客戶分類").sum().reset_index()
```

## 樞紐分析表

```python
df = pd.read_excel("C:/ACD019600/data/test10.xlsx")
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count')
```

```python
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count')
```

```python
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count',margins = True)
```

```python
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count',margins = True,margins_name = "總計")
```

```python
#將缺失值填充為0
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count',margins = True,fill_value = 0)
```

```python
pd.pivot_table(df,values = ["使用者ID","7月銷量"],columns = "區域",index = "客戶分類",aggfunc={"使用者ID":"count","7月銷量":"sum"})
```

```python
pd.pivot_table(df,values = "使用者ID",columns = "區域",index = "客戶分類",aggfunc='count')
```
