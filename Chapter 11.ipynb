## 表的橫向拼接
### 連接表的類型

```python
import pandas as pd
data1 = {"名次":[1,2,3,4],
         "姓名":["小張","小王","小李","小趙"],
         "學號":["100","101","102","103"],
         "成績":[650,600,578,550]}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {"學號":["100","101","102","103"],
         "班級":["一班","一班","二班","二班"]}
df2 = pd.DataFrame(data2)
df2
```

```python
pd.merge(df1,df2)
```

```python
data1 = {"姓名":["小張","小王","小李"],
         "學號":["100","101","102"],
         "f_成績":[650,600,578]}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {"學號":['100',"100","101","102","101","102"],
         "e_成績":[586,602,691,702,645,676]}
df2 = pd.DataFrame(data2)
df2
```

```python
pd.merge(df1,df2,on = "學號")
```

### 連接方式

```python
data1 = {"名次":[1,2,3,4],
         "姓名":["小張","小王","小李","小趙"],
         "學號":["100","101","102","103"],
         "成績":[650,600,578,550]}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {"姓名":["小張","小王","小李","小錢"],
         "學號":["100","101","102","104"],
         "班級":["一班","一班","二班","三班"]}
df2 = pd.DataFrame(data2)
df2
```

```python
pd.merge(df1,df2,on = "學號",how = "inner")
```

```python
pd.merge(df1,df2,on = "學號",how = "left")
```

```python
pd.merge(df1,df2,on = "學號",how = "right")
```

```python
pd.merge(df1,df2,on = "學號",how = "outer")
```
